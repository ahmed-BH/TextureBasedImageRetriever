```python
import numpy as np
import matplotlib.pyplot as plt
import settings
from   cbir import TBIR
from   PIL  import Image
```

```python
my_cbir    = TBIR(dataset_dir=settings.DATASET_DIR, descriptors_dir=settings.DESCRIPTORS_DIR)
# my_cbir.offline_phase()
similar    = my_cbir.online_phase("dataset/wood/black-wattle-44x60.jpg", nb_results=28)

dpi        = 80
rows, cols = 4, int(28 / 4)
fig, axs   = plt.subplots(nrows=rows, ncols=cols, figsize=(15, 10), dpi=80, sharex=False, sharey=False)
for r in range(rows):
    for c in range(cols):
        img = Image.open(similar[c + r*cols]["image_path"], 'r')
        axs[r, c].imshow(img)
        axs[r, c].set_title(int(similar[c + r*cols]["distance"]))
```
